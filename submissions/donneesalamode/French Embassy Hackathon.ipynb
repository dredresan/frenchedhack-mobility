```python
import numpy as np
import pandas as pd
```

```python
visa_data = pd.read_excel('VISA_modified.xlsx')
```

```python
visa_data.columns = visa_data.columns.str.strip()
visa_data.columns = [c.replace(' ', '_') for c in visa_data.columns]
visa_data.columns = [x.lower() for x in visa_data.columns]
n_students = visa_data.shape[0]
print("Total Visa Students: " + str(n_students))

visa_americans_data = visa_data[visa_data['nationality'] == 'Etats-Unis']
n_americans = visa_americans_data.shape[0]
print("Total Americans: " + str(n_americans))
print("Percentage of Students that are American: " + str(round((n_americans/n_students),2) * 100) + "%")
visa_americans_data
```

```python
n_fem_americans = visa_americans_data[visa_americans_data['sexe'] == 'Féminin'].shape[0]
n_male_americans = visa_americans_data[visa_americans_data['sexe'] == 'Masculin'].shape[0]
print("Male Americans: " + str(n_male_americans))
print("Female Americans: " + str(n_fem_americans))
print("Percentage of Female Americans: " + str(round((n_fem_americans/n_americans),2) * 100) + "%")
```

```python
n_pos_americans = visa_americans_data[visa_americans_data['positive_avis_scac'] == 1].shape[0]
n_neg_americans = visa_americans_data[visa_americans_data['negative_avis_scac'] == 1].shape[0]
print("Postive Opinions by Americans: " + str(n_pos_americans))
print("Negative Opinions by Americans: " + str(n_neg_americans))
print("Percentage of Positive Opinions by Americans: " + str(round((n_pos_americans/n_americans),2) * 100) + "%")
```

```python
n_echange_amer = visa_americans_data[visa_americans_data['mobility_type'] == 'Echange'].shape[0]
n_accep_amer = visa_americans_data[visa_americans_data['mobility_type'] == 'Acceptation papier'].shape[0]
print("Num of 'Echange' Americans: " + str(n_echange_amer))
print("Num of 'Acceptation papier' Americans: " + str(n_accep_amer) + "\n")

print("Percentage of 'Echange' Americans: " + str(round((n_echange_amer/n_americans),2) * 100) + "%")
print("Percentage of 'Acceptation papier' Americans: " + str(round((n_accep_amer/n_americans),2) * 100) + "%")
visa_americans_data.mobility_type.unique()
```
